### Creating Data

```python
# Mixture of Gaussian and log-Gaussian
z = np.random.uniform(0, 1, N) < 0.6 # assume mixture coefficients are (0.8,0.2)
dat = np.zeros(N)
for i in range(N):
    if z[i] == True:
        dat[i] = np.random.normal(1.5, 0.5, 1)
    else:
        dat[i] = np.random.lognormal(1, 0.25, 1)
        
noise = np.random.normal(0, 0.01, N) 
dat += noise

dat = np.c_[dat]
```

### aPCE

```python
def exp(X):
    return np.c_[np.exp(-X)]

mod = aPCE_withQuad(dat, p, exp)
pce_coef_Quad = mod.Quadrature(n_quad)
pce_coef_LR = mod.LR()
```

```python
def evaluate_aPCE_exp(pce_coef, d, max_deg, x):
    """
    evaluate pce expansion with maximum degree max_deg at the points x
    for Black-Scholes
    input:
    pce_coef: 1d array, the pce coefficients
    d: int, the number of parameter
    max_deg: int, the maximum degree
    x: 2d array, the points where the pce is evaluated. x.shape[1] = 3
    return
    val: 1d array of the same size as x.shape[0]
    """
    pce_index = mod.multivariate_pce_index(d, max_deg)
    val = np.zeros(x.shape[0])
    for i in range(pce_index.shape[0]):    
        val += pce_coef[i]*\
               (mod.Pol_eval(mod.Create_Orthonormal_Polynomials('Quadrature')[0][0][pce_index[i][0]], x)).reshape(N)
    return val
```

```python
def evaluate_aPCE_exp2(pce_coef, d, max_deg, x):
    """
    evaluate pce expansion with maximum degree max_deg at the points x
    for Black-Scholes
    input:
    pce_coef: 1d array, the pce coefficients
    d: int, the number of parameter
    max_deg: int, the maximum degree
    x: 2d array, the points where the pce is evaluated. x.shape[1] = 3
    return
    val: 1d array of the same size as x.shape[0]
    """
    pce_index = mod.multivariate_pce_index(d, max_deg)
    val = np.zeros(x.shape[0])
    for i in range(pce_index.shape[0]):    
        val += pce_coef[i]*\
               (mod.Pol_eval(mod.Create_Orthonormal_Polynomials('LR')[0][pce_index[i][0]], x)).reshape(N)
    return val
```

```python
y_aPCE_Quad = evaluate_aPCE_exp(pce_coef_Quad, d, p, dat2)
y_aPCE_LR = evaluate_aPCE_exp2(pce_coef_LR, d, p, dat2)

y_true = exp(dat2)
```

## Ishigami Function

```python
N = 300 # Sample size
N_test = 100
d = 3  # d --> number of random variables
p = 12 # p --> degree of PCE polynomial
n_quad = 14
n = int(math.factorial(d + p)/(math.factorial(d)*math.factorial(p))) # n --> Number of terms in the PCE expansion
```

### Data

```python
np.random.seed(100)
#noise = np.random.normal(0, 0.01, N) 

#X = np.random.uniform(-1,1, size=(N, d)) + np.c_[noise]
X = np.random.uniform(-1,1, size=(N, d))
X_test = np.random.uniform(-1,1, size=(N_test, d))
```

```python
X *= np.pi
X_test *= np.pi
```

```python
def multivariate_pce_index(d, max_deg):
    """
    Generate all the d-dimensional polynomial indices with the 
    constraint that the sum of the indexes is <= max_deg

    input:
    d: int, number of random variables
    max_deg: int, the max degree allowed

    return: 
    2d array with shape[1] equal to d, the multivariate indices
    """
    maxRange = max_deg*np.ones(d, dtype = 'int')
    index = np.array([i for i in product(*(range(i + 1) for i in maxRange)) if sum(i) <= max_deg])

    return index
```

```python
idx = multivariate_pce_index(d, p)
```

```python
def Ishigami_Function(dat):
    Z = dat
    return np.c_[np.sin(Z[:,0]) + 7*(np.sin(Z[:,1]))**2 + 0.1*Z[:,2]**4*np.sin(Z[:,0])]
```

```python
Y = Ishigami_Function(X)
```

### aPCE

```python
mod = aPCE_withQuad(X, Ishigami_Function, p, idx)
pce_coef_Quad, P_quad = mod.Quadrature(n_quad)
pce_coef_LR, P_LR = mod.LR()
```

```python
def evaluate_aPCE_Ishigami(pce_coef, P, x, idx):
    """

    """
    val = np.zeros(x.shape[0])
    for i in range(idx.shape[0]):    
        val += pce_coef[i]*\
               mod.Pol_eval(P[0][idx[i,0],:], x[:,0])*\
               mod.Pol_eval(P[1][idx[i,1],:], x[:,1])*\
               mod.Pol_eval(P[2][idx[i,2],:], x[:,2])
    return val
```

```python
y_aPCE_Quad_Train = evaluate_aPCE_Ishigami(pce_coef_Quad, P_quad, X, idx)
y_aPCE_Quad_Test = evaluate_aPCE_Ishigami(pce_coef_Quad, P_quad, X_test, idx)
y_aPCE_LR = evaluate_aPCE_Ishigami(pce_coef_LR, P_LR, X_test, idx)

y_true = Ishigami_Function(X_test)
```

## PCE

```python
def basis_pce(Z, p):
    d = Z.shape[1]
    N = Z.shape[0]
    n = int(math.factorial(d + p)/(math.factorial(d)*math.factorial(p)))
        
    idx = multivariate_pce_index(d, p)
    Phi = np.ones((N, n))
    
    for i in range(n):
        for j in range(d):
            Phi[:,i] *=  math.sqrt(2*idx[i][j]+1)*legendre(idx[i][j])(Z[:,j]/np.pi)
    return Phi
```

## aPCE and AFVB method from Bhattacharyya

```python
## These are the optimal hyper-parameters found via GridSearchCV
# {'A_0': 1.0, 'B_0': 1e-05, 'C_0': 0.45, 'D_0': 0.00012}
```

```python
################
#### Inputs ####
################
n = int(math.factorial(d + p)/(math.factorial(d)*math.factorial(p))) # n --> Number of terms in the PCE expansion
# A_0 and B_0 --> the parameters of the Gamma distribution in the joint Gaussian-Gamma 
# distribution used to express the prior
A_0 = 0.01 
B_0 = 0.0001 
# C_0 and D_0 --> the paramters of the Gamma distribution in the distribution of the hyper-prior
C_0 = 0.01
D_0 = 0.0001 
T_ell = 0.001 # T_ell --> threshold value for the VLB (for the error e_ell) in percent
e = 1000 # e --> ddjustable parameter for the threshold value 
```

```python
def basis(Z, p):
    d = Z.shape[1]
    N = Z.shape[0]
    n = int(math.factorial(d + p)/(math.factorial(d)*math.factorial(p)))
        
    idx = multivariate_pce_index(d, p)

    Phi = np.ones((N, n))
    mod = aPCE(X, Ishigami_Function, p, idx)
    P = mod.Create_Orthonormal_Polynomials(p)
    for i in range(n):
        for j in range(d):
            Phi[:,i] *=  mod.Pol_eval(P[j][idx[i][j],:], Z[:,j])
    return Phi
```

## PCE Using Bhattacharyya Algorithm

```python
def basis_pce(Z, p):
    d = Z.shape[1]
    N = Z.shape[0]
    n = int(math.factorial(d + p)/(math.factorial(d)*math.factorial(p)))
        
    idx = multivariate_pce_index(d, p)
    Phi = np.ones((N, n))
    
    for i in range(n):
        for j in range(d):
            Phi[:,i] *=  math.sqrt((2*idx[i][j]+1)/1)*legendre(idx[i][j])(Z[:,j]/np.pi)
    return Phi
```

## GSA

```python
def GSA(a, idx, active_cols):
    
    V = np.sum(a[1:]**2)
    V_Partial = []
    V_Total = []
    S_Partial = []
    S_Total = []
    
    l = [0,1,2]
    
    for k in range(3):
        l = [0,1,2]
        l.remove(k)
        
        temp = np.full(idx[active_cols].shape[0], True, dtype = bool)
        
        for i in range(idx[active_cols].shape[0]):
            temp[i] = (idx[active_cols][i][k] != 0) & (sum(idx[active_cols][i][l]) == 0)
            
        V_Partial.append(np.sum(a[temp,:]**2))
        S_Partial.append(np.sum(a[temp,:]**2)/V)

        for i in range(idx[active_cols].shape[0]):
            temp[i] = idx[active_cols][i][k] != 0

        V_Total.append(np.sum(a[temp,:]**2))
        S_Total.append(np.sum(a[temp,:]**2)/V)

    
    
    return V, V_Partial, V_Total, S_Partial, S_Total
```

```python
print('GSA Results for AFVB_aPCE')
GSA(a_hat, idx, active_cols)
```

```python
print('GSA Results for VRVM_aPCE')
GSA(np.c_[c_sol[active_indices,0]], idx, active_indices)
```

```python
print('GSA Results for VRVM_PCE')
GSA(np.c_[c_sol_pce[active_indices_pce,0]], idx, active_indices_pce)
```

### 6.2 Sobol' function

The Sobol function is given by

$$
Y = \prod_{i=1}^d\frac{|4\xi_i - 2| + a_i}{1 + a_i}
$$

where $\xi_i$ are independent random variables uniformly distributed over $[0,1]$ and $a_i$ are the constants. 


```python
alpha = np.c_[1, 2, 5, 10, 20, 50, 100, 500]
```

```python
##########################################################################################
#### Generate N samples for d random variables and compute the responses at N samples ####
##########################################################################################
np.random.seed(10)
Z = np.zeros((N, d))
for i in range(d):
    Z[:,i] = np.random.uniform(-1,1,N)


Y = np.c_[np.prod((np.abs(4*(0.5*Z+1/2) - 2) + alpha)/(1 + alpha),1)]
```
